```python
import pandas as pd
import matplotlib.pyplot as plt
import functools
import numpy as np
```

```python
directory = 'D:/Data Science Projects/Malaysia Statistics/01.Principal_Statistics.xlsx'
```

```python
df_malaysia = pd.read_excel(directory, sheet_name=0)
df_johor = pd.read_excel(directory, sheet_name=1)
df_kedah = pd.read_excel(directory, sheet_name=2)
df_kelantan = pd.read_excel(directory, sheet_name=3)
df_melaka = pd.read_excel(directory, sheet_name=4)
df_nsembilan = pd.read_excel(directory, sheet_name=5)
df_pahang = pd.read_excel(directory, sheet_name=6)
df_ppinang = pd.read_excel(directory, sheet_name=7)
df_perak = pd.read_excel(directory, sheet_name=8)
df_perlis = pd.read_excel(directory, sheet_name=9)
df_selangor = pd.read_excel(directory, sheet_name=10)
df_terangganu = pd.read_excel(directory, sheet_name=11)
df_sabah = pd.read_excel(directory, sheet_name=12)
df_sarawak = pd.read_excel(directory, sheet_name=13)
df_kl = pd.read_excel(directory, sheet_name=14)
df_labuan = pd.read_excel(directory, sheet_name=15)
df_putrajaya = pd.read_excel(directory, sheet_name=16)
```

```python
df_malaysia
```

<h1>Total of Unemployment (thousands) in Malaysia 1982-2020

```python
plt.figure(figsize=(16,8))
plt.title('Total of Unemployment (thousands) in Malaysia 1982-2020')
plt.plot(df_malaysia['Year'], df_malaysia['Unemployed'])
plt.xlabel('Year')
plt.ylabel('Number of Unemployed'+ ' (thousands)')
plt.show()
```

<h1>Rate of Unemployment (%) in Malaysia 1982-2020

```python
plt.figure(figsize=(16,8))
plt.title('Rate of Unemployment (%) in Malaysia 1982-2020')
plt.plot(df_malaysia['Year'], df_malaysia['Unemployment rate'])
plt.xlabel('Year')
plt.ylabel('Rate of Unemployment (%)')
plt.show()
```

<h1>Total of Unemployment (thousands) in Malaysia 1982-2020 by States

```python
plt.figure(figsize=(16,8))
plt.title('Total of Unemployment (thousands) in Malaysia 1982-2020 by States')
plt.plot(df_johor['Year'], df_johor['Unemployed'])
plt.plot(df_kedah['Year'], df_kedah['Unemployed'])
plt.plot(df_kelantan['Year'], df_kelantan['Unemployed'])
plt.plot(df_melaka['Year'], df_melaka['Unemployed'])
plt.plot(df_nsembilan['Year'], df_nsembilan['Unemployed'])
plt.plot(df_pahang['Year'], df_pahang['Unemployed'])
plt.plot(df_ppinang['Year'], df_ppinang['Unemployed'])
plt.plot(df_perak['Year'], df_perak['Unemployed'])
plt.plot(df_perlis['Year'], df_perlis['Unemployed'])
plt.plot(df_selangor['Year'], df_selangor['Unemployed'])
plt.plot(df_terangganu['Year'], df_terangganu['Unemployed'])
plt.plot(df_sabah['Year'], df_sabah['Unemployed'])
plt.plot(df_sarawak['Year'], df_sarawak['Unemployed'])
plt.plot(df_kl['Year'], df_kl['Unemployed'])
plt.plot(df_labuan['Year'], df_labuan['Unemployed'])
plt.plot(df_putrajaya['Year'], df_putrajaya['Unemployed'])

plt.xlabel('Year')
plt.ylabel('Number of Unemployed'+ ' (thousands)')
plt.legend()
plt.show()
```

```python
dfs = [df_johor[['Year', 'Unemployed']],
df_kedah[['Year', 'Unemployed']],
df_kelantan[['Year', 'Unemployed']],
df_melaka[['Year', 'Unemployed']],
df_nsembilan[['Year', 'Unemployed']],
df_pahang[['Year', 'Unemployed']],
df_ppinang[['Year', 'Unemployed']],
df_perak[['Year', 'Unemployed']],
df_perlis[['Year', 'Unemployed']],
df_selangor[['Year', 'Unemployed']],
df_terangganu[['Year', 'Unemployed']],
df_sabah[['Year', 'Unemployed']],
df_sarawak[['Year', 'Unemployed']],
df_kl[['Year', 'Unemployed']],
df_labuan[['Year', 'Unemployed']],
df_putrajaya[['Year', 'Unemployed']]
]
```

```python
df_unemployed_by_states = functools.reduce(lambda left,right: pd.merge(left, right, on='Year', how='outer',), dfs)
df_unemployed_by_states.columns = ['Year',
'Johor',
'Kedah',
'Kelantan',
'Melaka',
'NSembilan',
'Pahang',
'PPinang',
'Perak',
'Perlis',
'Selangor',
'Terangganu',
'Sabah',
'Sarawak',
'KL',
'Labuan',
'Putrajaya'
]
```

```python
df_unemployed_by_states
```

```python
plt.figure(figsize=(16,8))
plt.title('Total of Unemployment '+'(thousands)'+' in Malaysia 1985-2020')
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Johor'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Kedah'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Kelantan'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Melaka'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['NSembilan'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Pahang'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['PPinang'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Perak'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Perlis'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Selangor'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Terangganu'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Sabah'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Sarawak'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['KL'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Labuan'])
plt.plot(df_unemployed_by_states['Year'], df_unemployed_by_states['Putrajaya'])
plt.xlabel('Year')
plt.ylabel('Number of Unemployed (thousands)')
plt.legend(['Johor',
            'Kedah',
            'Kelantan',
            'Melaka',
            'NSembilan',
            'Pahang',
            'PPinang',
            'Perak',
            'Perlis',
            'Selangor',
            'Terangganu',
            'Sabah',
            'Sarawak',
            'KL',
            'Labuan',
            'Putrajaya'])
plt.show()
```
